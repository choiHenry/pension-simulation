```python
import pandas as pd
```

```python
wage_function = pd.read_csv("../data/new/wage_function.txt", sep=" ")
```

```python
wage_function
```

```python
wage_function.shape
```

```python
wage_function['yos'] -= 1
```

```python
wage_history = pd.DataFrame()
wage_history[['group_4', 'group_5', 'group_6']] = table.iloc[:, -3:] * (1.021)**3
```

```python
wage_history['group_1'] = wage_function['group_1'] * base_wage[0]
```

```python
wage_history['group_2'] = wage_function['group_2'] * base_wage[1]
```

```python
wage_history['group_3'] = wage_function['group_3'] * base_wage[2]
```

```python
wage_history
```

```python
import numpy as np
wage_history['group_1'] = wage_history['group_1'].shift(periods=3, fill_value=np.nan) 
```

```python
wage_history
```

```python
wage_history.loc[36:, 'group_1'] = np.nan
```

```python
wage_history
```

```python
wage_history['group_2'] = wage_history['group_2'].shift(periods=7, fill_value=np.nan) 
wage_history.loc[37:, 'group_2'] = np.nan
wage_history
```

```python
wage_history['group_3'] = wage_history['group_3'].shift(periods=5, fill_value=np.nan) 
wage_history.loc[36:, 'group_3'] = np.nan
wage_history
```

```python
wage_history = wage_history[['group_1', 'group_2', 'group_3', 'group_4', 'group_5', 'group_6']]
wage_history.index.name = 'yos'
wage_history
```

```python
wage_history_1970 = pd.DataFrame()
wage_history_1970['year'] = pd.DataFrame(range(1989, 2035))
```

```python
wage_history_1970['age'] = pd.DataFrame(range(19, 65))
wage_history_1970[['group_1', 'group_2', 'group_3', 'group_4', 'group_5', 'group_6']] = wage_history[['group_1', 'group_2', 'group_3', 'group_4', 'group_5', 'group_6']]
```
