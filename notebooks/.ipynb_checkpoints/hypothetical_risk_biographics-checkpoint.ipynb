```python
# 기업규모 변수 구성2
def fsize2(row):
    import numpy as np
    temp = row['fsize']
    if np.isnan(temp):
        return
    if temp >= 1 and temp <= 4:
        return '소기업'
    elif temp >= 5 and temp <= 7:
        return '중기업'
    elif temp >= 8:
        return '대기업'
df22p['fsize2'] = df22p.apply(fsize2, axis=1)
# df22p.apply(fsize, axis=1)
```

```python
# 기업규모 변수 구성
def jtype(row):
    import numpy as np
    temp = row['p220351']
    if np.isnan(temp):
        return
    if temp >= 100 and temp < 400:
        return '화이트 칼라'
    elif temp >= 600 and temp < 900:
        return '블루 칼라'
    else:
        print(temp)
        return 'not classified'
df22p['jtype'] = df22p.apply(jtype, axis=1)
# df22p.apply(fsize, axis=1)
```

```python
# 기업규모 변수 구성
def jtype2(row):
    import numpy as np
    temp = row['p220314']
    if np.isnan(temp):
        return
    if temp == 1:
        return '상용직'
    elif temp >= 2 and temp <= 3:
        return '임시일용직'
    else:
        print(temp)
        return 'not classified'
df22p['jtype2'] = df22p.apply(jtype2, axis=1)
# df22p.apply(fsize, axis=1)
```

```python
# 기업규모 변수 구성
def sex(row):
    import numpy as np
    temp = row['p220101']
    if np.isnan(temp):
        return
    if temp == 1:
        return '남성'
    elif temp >= 2 and temp <= 3:
        return '여성'
    else:
        print(temp)
        return 'not classified'
df22p['성별'] = df22p.apply(sex, axis=1)
# df22p.apply(fsize, axis=1)
```
